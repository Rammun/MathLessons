### Polar coordinates

```python
def polar_to_dec(r, phi):
    x = r * np.cos(phi)
    y = r * np.sin(phi)
    return x, y

coordinates = polar_to_dec(1, 30)
print(coordinates)
```

```python
phi = np.linspace(0, 2 * np.pi, 100)
R = 5

plt.plot(R * np.sin(phi), R * np.cos(phi))

plt.axis('equal')
plt.grid(True)
plt.show()
```
