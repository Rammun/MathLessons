## Task 3

```python
t = np.linspace(0, 2 * np.pi, 100)
R = 4

plt.plot(R * np.sin(t), R * np.cos(t))
plt.axis('equal')
plt.show()
```

```python
r1 = 4
r2 = 2

t = np.linspace(0, 2 * np.pi, 100)
plt.plot(r1 * np.cos(t), r2 * np.sin(t))
plt.axis('equal')
plt.show()
```

```python
k = 10

x = np.linspace(0, 20, 100)
mx = ma.masked_values (x, 0)

plt.plot(mx, k / mx)
plt.show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)
z = 2 * x + 3 * y
d = 30
ax.plot_wireframe(x, y, z)
ax.plot_wireframe(x, y, z + d)

show()
```

```python
x = np.arange(-5, 5, 0.5)
y = np.arange(-5, 5, 0.5)
x, y = np.meshgrid(x, y)

fig = figure()
ax = Axes3D(fig)
z = x**2 + y**2
ax.plot_wireframe(x, y, z)
show()

fig = figure()
ax = Axes3D(fig)
z = x**3 + y**3
ax.plot_wireframe(x, y, z + d)
show()
```
