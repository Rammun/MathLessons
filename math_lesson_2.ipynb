```python
t = np.linspace(0, 2 * np.pi, 100)
R = 4

plt.plot(R * np.sin(t), R * np.cos(t))
plt.axis('equal')
plt.show()
```

```python
r1 = 4
r2 = 2

t = np.linspace(0, 2 * np.pi, 100)
plt.plot(r1 * np.cos(t), r2 * np.sin(t))
plt.axis('equal')
plt.show()
```
